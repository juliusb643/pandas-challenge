```python
import pandas as pd
from pathlib import Path

school_data_to_load = Path("C:\\Users\juliu\OneDrive\Desktop\penn-resource-files\Module 4 Schools Project\Mod-4-Starter-Code\PyCitySchools\Resources\schools_complete.csv")
student_data_to_load = Path("C:\\Users\juliu\OneDrive\Desktop\penn-resource-files\Module 4 Schools Project\Mod-4-Starter-Code\PyCitySchools\Resources\students_complete.csv")

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
school_count = school_data_complete["school_name"].drop_duplicates().count()
school_count
```

```python
student_count = school_data_complete["Student ID"].count()
student_count
```

```python
total_budget = school_data_complete["budget"].drop_duplicates().sum()
total_budget
```

```python
district_average_math_score = school_data_complete["math_score"].mean()
district_average_math_score
```

```python
district_average_reading_score = school_data_complete["reading_score"].mean()
district_average_reading_score
```

```python
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage 
```

```python
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

## District Summary DataFrame

```python
district_summary = pd.DataFrame({
    'Total Schools': [school_count],
    'Total Students': [student_count],
    'Total Budget': [total_budget],
    'Average Math Score': [district_average_math_score],
    'Average Reading Score': [district_average_reading_score],
    '% Passing Math': [passing_math_percentage],
    '% Passing Reading': [passing_reading_percentage],
    '% Overall Passing': [overall_passing_rate]})

district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary = district_summary.reset_index(drop=True)
district_summary
```

```python
school_types = school_data_complete.groupby("school_name")["type"].first()
#school_types.index.name = 'School Name'
school_types
```

```python
per_school_count = school_data_complete.groupby("school_name")["student_name"].count()
#per_school_count.index.name = 'School Name'
per_school_count
```

```python
per_school_budget = school_data_complete.groupby("school_name")["budget"].first()
#per_school_budget.index.name = 'School Name'
per_school_budget
```

```python
per_school_capita = per_school_budget / per_school_count
#per_school_capita.index.name = 'School Name'
per_school_capita
```

```python
school_average_math_score = school_data_complete.groupby("school_name")
school_average_math_score = school_average_math_score["math_score"].mean()
#school_average_math_score.index.name = 'School Name'
school_average_math_score
```

```python
school_average_reading_score = school_data_complete.groupby("school_name")
school_average_reading_score = school_average_reading_score["reading_score"].mean()
#school_average_reading_score.index.name = 'School Name'
school_average_reading_score
```

```python
school_passing_math_count = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name").size()
student_count = school_data_complete.groupby("school_name")["student_name"].count()
passing_math_percentage = (school_passing_math_count / student_count) * 100
#passing_math_percentage.index.name = "School Name"
passing_math_percentage 
```

```python
school_passing_reading_count = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name").size()
student_count = school_data_complete.groupby("school_name")["student_name"].count()
passing_reading_percentage = (school_passing_reading_count / student_count) * 100
#passing_reading_percentage.index.name = "School Name"
passing_reading_percentage 
```

```python
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
school_students_passing_math_and_reading = school_students_passing_math_and_reading / per_school_count * 100
#school_students_passing_math_and_reading.index.name = "School Name"
school_students_passing_math_and_reading
```

## Per School Summary DataFrame

```python
per_school_summary = pd.DataFrame({
    'School Type': school_data_complete.groupby("school_name")["type"].first(),
    'Total Students': per_school_count.groupby("school_name").first(),
    'Total School Budget': per_school_budget,
    'Per Student Budget': per_school_capita,
    'Average Math Score': school_average_math_score.groupby("school_name").first(),
    'Average Reading Score': school_average_reading_score.groupby("school_name").first(),
    '% Passing Math': passing_math_percentage.groupby("school_name").first(),
    '% Passing Reading': passing_reading_percentage.groupby("school_name").first(),
    '% Overall Passing': school_students_passing_math_and_reading.groupby("school_name").first()})

per_school_summary.index.name = None
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

per_school_summary
```

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["0-$585", "$585-630", "$630-645", "$645-680"]
school_spending_df = per_school_summary.copy()
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].str.replace("$","").astype("float")
school_spending_df['Spending Range'] = pd.cut(school_spending_df["Per Student Budget"], bins=spending_bins, labels=labels)
school_spending_df
```

```python
spending_math_scores = school_spending_df.groupby(["Spending Range"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Range"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Range"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Range"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Range"])["% Overall Passing"].mean()
```

## Spending Summary DataFrame

```python
spending_summary= pd.DataFrame({
    'Spending Average Math': spending_math_scores,
    'Spending Average Reading': spending_reading_scores,
    'Spending Passing Math': spending_passing_math,
    'Spending Passing Reading': spending_passing_reading,
    'Spending Passing Overall': overall_passing_spending})

spending_summary
```

```python
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
size_summary = per_school_summary.copy()
size_summary['Size Category'] = pd.cut(size_summary['Total Students'], bins=size_bins, labels=labels)
size_summary
```

```python
category_math_scores = size_summary.groupby(["Size Category"])["Average Math Score"].mean()
category_reading_scores = size_summary.groupby(["Size Category"])["Average Reading Score"].mean()
category_passing_math = size_summary.groupby(["Size Category"])["% Passing Math"].mean()
category_passing_reading = size_summary.groupby(["Size Category"])["% Passing Reading"].mean()
category_overall_passing = size_summary.groupby(["Size Category"])["% Overall Passing"].mean()
```

## Size Summary DataFrame

```python
size_summary_by_category= pd.DataFrame({
    'Category Math Scores': category_math_scores,
    'Category Reading Scores': category_reading_scores,
    'Category Passing Math': category_passing_math,
    'Category Passing Reading': category_passing_reading,
    'Category Overall Passing': category_overall_passing})

size_summary_by_category
```
